```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
datapath = path + 'AE 460: Aerodynamics and Propulsion/AE 460 Lab Group/AE 460 Lab 4/AE 460 Lab 4 Data/hotwire calibration.csv'
df = pd.read_csv(datapath)
df
```

```python
x = df['Hotwire Ev [Volts]']
y1 = df['Pitot/static Probe Velocity [m/s]']
y2 = -9.368 + -5.803*x + 35.304*x**2 + -31.765*x**3 + 9.296*x**4
plt.title("Hotwire Ev vs Pitot/Static Probe Velocity and Calculated Velocity from Calibration Coeeficents")
plt.xlabel("Hotwire Ev [Volts]")
plt.ylabel("Velocity [m/s]")
plt.plot(x, y1, label = "Measured Velocity")
plt.plot(x, y2, linestyle=':', linewidth=3, label = "Calculated Velocity")
plt.legend()
plt.grid()
plt.show()
```
